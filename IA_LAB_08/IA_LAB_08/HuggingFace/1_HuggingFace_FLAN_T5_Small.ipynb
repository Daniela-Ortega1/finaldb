```python
from transformers import pipeline

# Creamos un pipeline de generación texto-a-texto con FLAN-T5-Small
qa_pipeline = pipeline(
    "text2text-generation",
    model="google/flan-t5-small"
)
```

```python
pregunta_1 = "Explain in simple terms what overfitting is in machine learning."
respuesta_1 = qa_pipeline(pregunta_1, max_new_tokens=128)[0]["generated_text"]

print("PREGUNTA 1:")
print(pregunta_1)
print("\nRESPUESTA 1:")
print(respuesta_1)
```

```python
pregunta_2 = (
    "Write a short motivational message for engineering students working "
    "on the MORITA project for blackberry pulp production."
)
respuesta_2 = qa_pipeline(pregunta_2, max_new_tokens=128)[0]["generated_text"]

print("PREGUNTA 2:")
print(pregunta_2)
print("\nRESPUESTA 2:")
print(respuesta_2)
```
