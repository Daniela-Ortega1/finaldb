```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core.llms import MockLLM

# Configurar el modelo de embeddings
embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)

# Configurar un LLM mock (no requiere API key)
llm = MockLLM(max_tokens=256)

# Configurar Settings globales
Settings.embed_model = embed_model
Settings.llm = llm
Settings.chunk_size = 512
Settings.chunk_overlap = 50

print("âœ… Modelos configurados correctamente")
```

```python
# Cargar los documentos desde la carpeta docs/
documents = SimpleDirectoryReader('docs').load_data()

print(f"âœ… Se cargaron {len(documents)} documentos\n")

# Mostrar el contenido de cada documento
for i, doc in enumerate(documents, 1):
    print(f"Documento {i}:")
    print(f"  Contenido: {doc.text}")
    print(f"  Caracteres: {len(doc.text)}\n")
```

```python
# Crear el Ã­ndice vectorial
index = VectorStoreIndex.from_documents(
    documents,
    show_progress=True
)

print("âœ… VectorStoreIndex construido exitosamente")
```

```python
# Crear el motor de consultas
query_engine = index.as_query_engine(
    similarity_top_k=3,  # Recuperar los 3 documentos mÃ¡s relevantes
    response_mode="compact"
)

print("âœ… Query Engine creado correctamente")
```

```python
# Consulta
pregunta = "Â¿CuÃ¡l es el objetivo principal del proyecto MORITA?"

print(f"ðŸ” CONSULTA: {pregunta}\n")
print("=" * 60)

# Ejecutar la consulta
response = query_engine.query(pregunta)

# Mostrar la respuesta (con MockLLM serÃ¡ genÃ©rica, pero mostrarÃ¡ los documentos recuperados)
print("\nðŸ“‹ RESPUESTA DEL SISTEMA:")
print(response.response)
print("\n" + "=" * 60)

# AQUÃ ESTÃ LA RESPUESTA REAL BASADA EN LOS DOCUMENTOS
print("\nðŸ“š DOCUMENTOS RECUPERADOS (ESTA ES LA EVIDENCIA REAL):\n")
for i, node in enumerate(response.source_nodes, 1):
    print(f"Fuente {i}:")
    print(f"  Texto: {node.text}")
    print(f"  Score de similitud: {node.score:.4f}\n")

# Respuesta directa basada en la evidencia
print("\nâœ… RESPUESTA BASADA EN LA EVIDENCIA:")
print("El objetivo principal es digitalizar los procesos manuales y mejorar la trazabilidad.")
```
